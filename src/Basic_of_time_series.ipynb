Plotting the Data

```python
df_index.spx.plot(figsize=(20,5), title="SPX Prices")
plt.show()
```

```python
df_index.ftse.plot(figsize=(20,5), title="FTSE Prices")
plt.show()
```

```python
df_index.spx.plot(figsize=(20,5), title="SPX Prices")
df_index.ftse.plot(figsize=(20,5), title="FTSE Prices")
plt.title("SPX vs FTSE")
plt.show()
```

The QQ Plot

```python
scipy.stats.probplot(df_index.spx, plot=pylab)
pylab.show()
```
